```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 14,            # base font size
    "axes.labelsize": 16,       # x/y labels
    "axes.titlesize": 17,       # title
    "xtick.labelsize": 13,
    "ytick.labelsize": 13,
    "legend.fontsize": 13,
    "figure.dpi": 300
})

plt.figure(figsize=(8, 5))
```

## SVM

```python
# Loading the results
svm_incorrect = pd.read_csv("models/svm_incorrect.csv")

# For incorrect normalisation
list_best_rows = []
for method in svm_incorrect['Method'].unique():
  method_type = svm_incorrect[svm_incorrect['Method'] == method]
  for time in method_type['Time'].unique():
    subset = method_type[method_type['Time'] == time]
    best_tss_row = subset.loc[subset['TSS Mean'].idxmax()]
    list_best_rows.append(best_tss_row)
best_tss_df = pd.DataFrame(list_best_rows).reset_index(drop=True)

fig, axes = plt.subplots(2, 1, figsize=(8, 10), dpi = 300)

for method in best_tss_df['Method'].unique():
    subset = best_tss_df[best_tss_df['Method'] == method]
    axes[0].errorbar(subset['Time'], subset['TSS Mean'],yerr = subset['TSS Std'], marker='o', capsize=5, label=method)
    axes[1].errorbar(subset['Time'], subset['HSS Mean'],yerr = subset['HSS Std'], capsize=5, marker='o', label=method)

axes[0].set_title('TSS Mean vs Time')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('TSS Mean')
axes[0].set_xticks(range(12,121,12))
axes[0].set_ylim(0, 1)
axes[0].legend()

axes[1].set_title('HSS Mean vs Time')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('HSS Mean')
axes[1].set_xticks(range(12,121,12))
axes[1].set_ylim(0, 1)
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
svm_correct = pd.read_csv("models/svm_correct.csv")

# For correct normalisation
list_best_rows = []
for method in svm_correct['Method'].unique():
  method_type = svm_correct[svm_correct['Method'] == method]
  for time in method_type['Time'].unique():
    subset = method_type[method_type['Time'] == time]
    best_tss_row = subset.loc[subset['TSS Mean'].idxmax()]
    list_best_rows.append(best_tss_row)
best_tss_df = pd.DataFrame(list_best_rows).reset_index(drop=True)

fig, axes = plt.subplots(2, 1, figsize=(8, 10), dpi = 300)

for method in best_tss_df['Method'].unique():
    subset = best_tss_df[best_tss_df['Method'] == method]
    axes[0].errorbar(subset['Time'], subset['TSS Mean'], yerr = subset['TSS Std'], marker='o', capsize=5, label=method)
    axes[1].errorbar(subset['Time'], subset['HSS Mean'], yerr = subset['HSS Std'], marker='o', capsize=5, label=method)

axes[0].set_title('TSS Mean vs Time')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('TSS Mean')
axes[0].set_ylim(0, 1)
axes[0].set_xticks(range(12,121,12))
axes[0].legend()

axes[1].set_title('HSS Mean vs Time')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('HSS Mean')
axes[1].set_ylim(0, 1)
axes[1].set_xticks(range(12,121,12))
axes[1].legend()
axes[1].set_visible(False)

plt.tight_layout()
plt.show()
```
